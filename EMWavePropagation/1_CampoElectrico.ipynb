## 1.Ondas Electromagnéticas.

El campo eléctrico de una onda monocromática (de frecuencia única) puede describirse en función del tiempo $t$ y la distancia $z$ mediante la siguiente ecuación:

<br>
$$\large E(z,t) = E_0 \,cos(\omega t - \beta z) \, [V/m]$$<br>

donde:
- $E_0$ es la **amplitud** de la onda [V/m]. <br>
- $\omega$ es la frecuencia angular de la onda [rad/s]. <br>
- $\beta$ es la constante de fase de la onda [rad/m]. 

La **longitud de onda** $\lambda\$ de la onda se relaciona con la frecuencia a través de la siguiente ecuación:

<br>
$$\large v = \lambda f$$<br>

donde:
- $v$ es la **velocidad de propagación** de la onda. [V/m] <br>

Para una onda electromagnética que viaja en el vacío:

<br>
$$\large v = c = 3 \times 10^{8} \, [m/s]$$<br>


### 1.1. Onda en el tiempo

La frecuencia angular se calcula de la siguiente forma:

<br>
$$\large \omega = 2\pi f$$<br>

donde $f$ es la **frecuencia** de la onda [Hz].

### 1.2. Onda en el espacio

La constante de fase se calcula de la siguiente forma:

<br>
$$\large \beta = 2\pi / \lambda$$<br>

En la siguiente figura se observa como se propaga el campo eléctrico en el vacío, en función del tiempo y el espacio, al variar diferentes parámetros.

- ¿Qué ocurre con la longitud de onda al incrementar la frecuencia?
- ¿Qué ocurre con el periodo de la señal al incrementar la frecuencia?



```
import numpy as np 
import matplotlib.pyplot as plt

Amplitud_Eo = 1 #@param {type:"slider", min:0.1, max:1, step:0.1}
Frecuencia_f_MHz = 5 #@param {type:"slider", min:1, max:5, step:1}

f = Frecuencia_f_MHz*1e6
T = 1/f
v = 3e8
lambd = v/f
beta = 2*np.pi/lambd
omega = 2*np.pi*f

z = np.arange(0,500,1)
t = np.arange(0,16e-7,1e-9)

Ez = Amplitud_Eo*np.cos(beta*z)
Et = Amplitud_Eo*np.cos(omega*t)

rec_t = plt.Rectangle((T/1e-6, -1.2), T/1e-6, 2.4, color='#d187f5',alpha=0.3)
rec_z = plt.Rectangle((lambd, -1.2), lambd, 2.4, color='#d187f5',alpha=0.3)

fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].axis([0, 500, -1.2, 1.2])
ax[0].add_artist(rec_z)
ax[0].annotate('$\lambda$', xy=(1.4*lambd, 1.0), xytext=(1.4*lambd, 1.0), size = 16) 
ax[0].plot(z, Ez)
ax[0].set_xlabel('Distancia z (m)', fontsize=14)


ax[1].axis([0, 1.6, -1.2, 1.2])
ax[1].add_artist(rec_t)
ax[1].annotate('T', xy=(1.4*T/1e-6, 1.0), xytext=(1.4*T/1e-6, 1.0), size = 16) 
ax[1].plot(t/1e-6, Et)
ax[1].set_xlabel('Tiempo t ($\mu s$)', fontsize=14)

fig.show()
```

## 2.Campo Eléctrico.

La ecuación para el campo eléctrico incidente de una onda polarizada linealmente para medios dieléctricos y conductores puede expresarse como:

<br><br>
$$\large E(z,t) = e^{-\alpha z} cos(\omega t - \beta z) \, [V/m]$$<br>

donde:
- $\alpha$ es la constante de atenuación de la onda [1/m] <br>

Se observa que la 'envolvente' de la señal es $e^{-\alpha z}$, es decir, la 'envolvente' determina la amplitud de la señal coseno.

En la Figura se presenta la variación de la amplitud de la onda en función de la distancia, dependiendo del valor de la constante de atenuación $\alpha$.


```
alpha = 0.0011 #@param {type:"slider", min:0, max:0.002, step:0.0001}

f = 1e6
T = 1/f
v = 3e8
lambd = v/f
beta = 2*np.pi/lambd
omega = 2*np.pi*f
delta = 1/alpha

z = np.arange(0,1000,1)
Env = np.exp(-alpha*z)
CS = np.cos(beta*z)
Ez = Env*CS

rec_d = plt.Rectangle((0, -1.2), delta, 2.4, color='#d187f5',alpha=0.3)

fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].axis([0, 1000, -1.2, 1.2])
ax[0].plot(z, Env, 'r--',label='Envolvente')
ax[0].plot(z, CS, label='Coseno')
ax[0].set_xlabel('Distancia z (m)', fontsize=14)
ax[0].set_ylabel('E [V/m]', fontsize=14)
ax[0].legend(loc='lower right')

ax[1].axis([0, 1000, -1.2, 1.2])
ax[1].add_artist(rec_d)
ax[1].plot(z, Env, 'r--',label='Envolvente')
ax[1].plot([0, 1000], [0.37, 0.37], 'k--', lw=1, label='0.37')
ax[1].plot(z, Ez, label='Campo eléctrico', lw=3)
ax[1].set_xlabel('Distancia z (m)', fontsize=14)
ax[1].legend(loc='lower right')

fig.show()
```
