## 1.Modulación de intensidad.

La señal del laser, que puede ser representada por una señal senoidal, es modulada por un pulso, que generalmente se asume de forma Gausiana.

La señal de salida del láser se puede modelar como un fasor:

$$p(t) = Re\{A(t)e^{j \omega_c t}\}$$

donde:
- $A(t)$ es un pulso Gausiano.
- $\omega_c$ es la frecuencia angular de la señal del láser.

## 2.Pulso Gausiano.

La ecuación de un pulso Gausiano es la siguiente:

$$\large A(t) = A_0 \cdot exp \left(-\frac{1+jC}{2} \left( \frac{t}{T_0} \right)^2 \right)$$ <br>

donde:
- $A_0$ es la amplitud.
- $C$ es el chirp.
- $T_0$ es la mitad del ancho del pulso.

Se observa que la señal Pulso Gausiano es una señal compleja.

### Ancho medio del pulso Gausiano.

Para observar el ancho medio del pulso se asumen los siguientes valores:
- $A_0 = 1$ : amplitud.
- $C = 0$ : chirp.
- $T_0 = 1 s$ : es la mitad del ancho del pulso.

Al ser el chirp cero, $C=0$, la señal es real.

```
#@title Pulso Gausiano
import numpy as np
import matplotlib.pyplot as plt

C = 0
T0 = 1
t = np.arange(-4,4,0.05)
A = np.exp(-0.5*(1+C*1j)*(t/T0)*(t/T0))

fig,ax = plt.subplots(figsize=(5,5))
ax.plot(t,np.abs(A))
ax.grid()
ax.set_xlabel('Tiempo [s]')
ax.set_ylabel('Amplitud')
ax.set_title('Magnitud')
fig.show()
```

## 3.Señal modulada en intensidad.

Se graficarán diferentes formas de señal de salida del láser:

$$p(t) = Re\{A(t)e^{j \omega_c t}\}$$

Se asumen los siguientes valores:
- $A_0 = 1$ : amplitud.
- $T_0 = 1 s$ : es la mitad del ancho del pulso.
- $f=2 Hz$ : frecuencia del láser.

Se presentan gráficas con diferentes valores de chirp.

```
#@title Señal modulada en intensidad
f = 2
T0 = 1
t = np.arange(-4,4,0.0001)
port = np.exp(2*np.pi*f*t*1j)

fig,ax = plt.subplots(1,3,figsize=(15,5))
C = 0
A = np.exp(-0.5*(1+C*1j)*(t/T0)*(t/T0))
ax[0].plot(t,np.real(A*port))
ax[0].grid()
ax[0].set_xlabel('Tiempo [s]', fontsize=14)
ax[0].set_ylabel('Amplitud', fontsize=14)
ax[0].set_title('Pulso modulado C=0', fontsize=14)

C = 4
A = np.exp(-0.5*(1+C*1j)*(t/T0)*(t/T0))
ax[1].plot(t,np.real(A*port))
ax[1].grid()
ax[1].set_xlabel('Tiempo [s]', fontsize=14)
ax[1].set_title('Pulso modulado C=4', fontsize=14)

C = -4
A = np.exp(-0.5*(1+C*1j)*(t/T0)*(t/T0))
ax[2].plot(t,np.real(A*port))
ax[2].grid()
ax[2].set_xlabel('Tiempo [s]', fontsize=14)
ax[2].set_title('Pulso modulado C=-4', fontsize=14)

fig.show()
```
